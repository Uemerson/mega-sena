# Preparar o dataset

```python
df = pd.read_excel("mega_sena_asloterias_ate_concurso_2669_sorteio.xlsx")
# df = pd.read_excel("for-test.xlsx")
```

```python
# a partir da linha 6
df = df[6:]
```

```python
# renomeia as colunas do dataset
columns = ['concurso','data','bola1','bola2','bola3','bola4','bola5','bola6']

for index, column in enumerate(columns):
  df.rename(columns={df.columns[index]: column}, inplace=True)
```

# Quais os n(15) números sorteados com mais frequência ?

```python
n = 15
df_melted = pd.melt(df[['bola1','bola2','bola3','bola4','bola5','bola6']])
value_counts = df_melted['value'].value_counts()
df_results = pd.DataFrame({'Número': value_counts.index, 'Frequência': value_counts.values})

# Ordene pelo número que mais aparece e pela quantidade de vezes
df_results = df_results.sort_values(by=['Frequência', 'Número'], ascending=False)

# Obtenha os n(15) números mais frequentes
df_results.head(n)
```

# Quais os n(15) números sorteados com mais frequência no dia 31 do mês de dezembro?

```python
df_filter = df.copy()
df_filter['data'] = pd.to_datetime(df['data'], format='%d/%m/%Y')

filter_condition=(df_filter['data'].dt.day == 31) & (df_filter['data'].dt.month == 12)
df_filter=df[filter_condition]
```

```python
n = 15
df_melted = pd.melt(df_filter[['bola1','bola2','bola3','bola4','bola5','bola6']])
value_counts = df_melted['value'].value_counts()
df_results = pd.DataFrame({'Número': value_counts.index, 'Frequência': value_counts.values})

# Ordene pelo número que mais aparece e pela quantidade de vezes
df_results = df_results.sort_values(by=['Frequência', 'Número'], ascending=False)

# Obtenha os n(15) números mais frequentes
df_results.head(n)
```
